```python
import cv2, time, pandas
from datetime import datetime

# create a datafrme to hold time stamps
df = pandas.DataFrame(columns=["Start","End"])

first_frame = None

# list to hold statuses
status_list = [None, None]

# list to hold time stamps
times = []

# takes a number 0,1,2 (0 = first camer, 1 = second camera....) or video file path
video = cv2.VideoCapture(0)

# sets the pixel dimensions for the video capture
video.set(3,1920) 
video.set(4, 1080)




while True: 
    # create an object to save the video
    check, frame = video.read()
    # (check => boolean, frame => numpyarray)
    
    # the state of no new objects in frame
    status = 0
    
    # converts images from BGR to Gray
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # blurs image to remove noise
    gray = cv2.GaussianBlur(gray,(21,21),0)
    
    
    
    if first_frame is None:
        first_frame = gray
        continue
        # continue, meaning to the beginning of the loop
    
    # finding the diffence of pixels of 1st frame to each consecutive one
    delta_frame = cv2.absdiff(first_frame,gray)  
    thresh_frame = cv2.threshold(delta_frame, 55, 255, cv2.THRESH_BINARY)[1]
    thresh_frame = cv2.dilate(thresh_frame, None, iterations=2)
    
    # finds the contours of the objects in motion
    (cnts,_) = cv2.findContours(thresh_frame.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    
    for contour in cnts:
        if cv2.contourArea(contour) < 6500:
            # Adjust this number depending on the size of objects you want to detect
            continue
        
        # the state of new objects in frame
        status = 1    
        
        # creates a rectangle around moving objects
        (x,y,w,h) = cv2.boundingRect(contour)
        cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 3)
 
    

    # create a list of statuses 
    status_list.append(status)
    
    # to conserve memory only save the last two data points
    status_list = status_list[-2:]
    
    if status_list[-1] == 1 and status_list[-2] == 0:
        times.append(datetime.now()) 
    if status_list[-1] == 0 and status_list[-2] == 1:
        times.append(datetime.now())     
    
    # creates pop-up window of video feed
    cv2.imshow("Gray Frame",gray)   
    cv2.imshow("Delta Frame",delta_frame)
    cv2.imshow("Threshold Frame",thresh_frame)
    cv2.imshow("Color Frame", frame)
  
    # stops the video recording with key 'q'
    key = cv2.waitKey(1)
    if key==ord('q'):
        
        if status==1:
            times.append(datetime.now())
        # adds the time stamp(s) of when objects enter and leave the frame
        
        break
    

print(status_list)
print(times)

# adds each time stamps to a list
for t in range(0,len(times),2):
    df=df.append({"Start":times[t], "End":times[t+1]}, ignore_index=True)
    

# saves the time stamps list to a CSV file     
df.to_csv("Motion_Times.csv")    
    
# the cuts-off the video feed   
video.release()

# this codeblock allows to close the pop-up window with any key (on MAC OS)        
cv2.startWindowThread()
cv2.destroyAllWindows()
for i in range(2):
    cv2.waitKey(1)  
```
